```python
texto = "Bruce Lee[b] (nascido Lee Jun Fan;[c] 27 de novembro de 1940 – 20 de julho de 1973) foi um americano de Hong Kong artista marcial, ator, cineasta e filósofo. Foi o fundador da Jeet Kune Do, a artes marciais híbridas filosofia que foi formada a partir das experiências de Lee em combates desarmados e autodefesa eclético, Zen Budista e Taoísta philosophies—as uma nova escola de pensamento de artes marciais.[2][3] Com a carreira cinematográfica abrangendo Hong Kong e os Estados Unidos,[4][5][6] Lee é considerado como a primeira estrela de cinema chinesa global e um dos artistas marciais mais influentes na história do cinema.[7] Conhecido por seus papéis em cinco longas-metragens filmes de artes marciais, Lee é creditado por ajudar a popularizar filmes de artes marciais na década de 1970 e promover Cinema de ação de Hong Kong.[8][9] Nasceu em São Francisco e cresceu em Hong Kong britânica, Lee foi apresentado ao Indústria cinematográfica de Hong Kong como ator infantil por seu pai.[10] Sua experiência inicial em artes marciais incluiu Asa Chun (treinado sob Homem IP), tai chi, boxe (ganhando um torneio de boxe de Hong Kong) e frequente rua lutando (bairro e telhado lutas). Em 1959, Lee mudou-se para Seattle, onde se matriculou no Universidade de Washington em 1961.[11] Foi durante este tempo nos Estados Unidos que ele começou a considerar ganhar dinheiro ensinando artes marciais, mesmo que ele aspirava a ter uma carreira de ator. Ele abriu sua primeira escola de artes marciais, administrada em sua casa em Seattle. Depois de adicionar uma segunda escola Oakland, Califórnia, certa vez ele chamou atenção significativa em 1964 Campeonato Internacional de Karatê de Long Beach da Califórnia fazendo demonstrações e falando. Posteriormente, ele se mudou para Los Angeles para lecionar, incluindo seus alunos Chuck Norris, Sharon Tate, e Kareem Abdul-Jabbar.Seus papéis na América, inclusive atuando Kato em O Hornet Verde apresentou-o ao público americano, mas os atritos culturais que existiam não permitiram que Lee se expressasse plenamente. Depois de retornar a Hong Kong em 1971, Lee conseguiu seu primeiro papel principal O Grande Chefe, dirigido por Lo Wei. . Um ano depois, ele estrelou Punho da Fúria, em que ele retratou Chen Zhen, e O Caminho do Dragão, dirigido e escrito por Lee. Ele estrelou a coprodução EUA-Hong Kong Entra o Dragão (1973) e O Jogo da Morte (1978).[12] Seus filmes produzidos em Hong Kong e Hollywood, todos de sucesso comercial, elevaram os filmes de artes marciais de Hong Kong a um novo nível de popularidade e aclamação, provocando uma onda de interesse ocidental Artes marciais chinesas. . A direção e o tom de seus filmes, incluindo os deles coreografia luta e diversificação,[13] influenciou e mudou dramaticamente os filmes de artes marciais e artes marciais em todo o mundo.[14] Com a sua influência, kung fu filmes começou a deslocar o wuxia as lutas de gênero cinematográfico foram coreografadas de forma mais realista, os elementos de fantasia foram descartados para conflitos do mundo real e a caracterização do protagonista masculino passou de simplesmente um herói cavalheiresco para um que incorporasse a noção de masculinidade.[15]A carreira de Lee foi interrompida por sua morte repentina aos 32 anos edema cerebral, cujas causas continuam a ser uma questão controversa.[16] Mesmo assim, seus filmes permaneceram populares e ganharam muito culto seguinte,[8] e tornou-se amplamente imitado e explorado. . Ele se tornou um icônico figura conhecida em todo o mundo, principalmente entre os chineses, com base em sua representação de Cultura cantonesa em seus filmes,[17][página necessária] e entre os ásio-americanos por desafiarem Estereótipos asiáticos nos Estados Unidos.[18] Desde sua morte, Lee continuou a ser uma influência proeminente na modernidade esportes combate, incluindo judô, caratê, artes marciais mistas, e boxe, bem como a cultura popular moderna, incluindo cinema, televisão, quadrinhos, animação e videogames. Tempo nomeou Lee como um dos 100 pessoas mais importantes do século XX."
pergunta = "quem foi bruce lee?"
resposta = qea(question=pergunta, context=texto)
print("Pergunta: ", pergunta)
print("Resposta:",resposta['answer'])
print("Score:", resposta['score'])
```
