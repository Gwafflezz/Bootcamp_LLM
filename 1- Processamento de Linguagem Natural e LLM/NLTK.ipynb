```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer, SnowballStemmer, LancasterStemmer # Fixed: WordNetLemmatizer import corrected
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag, pos_tag_sents
import string

nltk.download('stopwords')
nltk.download('punkt')
nltk.download("tagsets")
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download("maxent_ne_chunker")
nltk.download("words")
nltk.download("punkt_tab")
nltk.download("tagsets_json")
nltk.download("averaged_perceptron_tagger_eng")
nltk.download("maxent_ne_chunker_tab")
```

```python
texto = "Nós somos feitos de poeira de estrelas. Nós somos uma maneira de o cosmos se autoconhecer. A imaginação nos leva a mundos que nunca sequer existiram. Mas sem ela não vmaos a lugar algum."
texto = ""
print(texto)
```

**Produção de tokens**

```python
nltk.download()
```

```python
sentencas = sent_tokenize(texto,language="portuguese")
print(type(sentencas))
print(sentencas)
```

imprime uma lista de senteças identificadas no corpus

```python
tokens = word_tokenize(texto,language="portuguese")
print(tokens)
print("quantidade de tokens:",len(tokens))
```

**Stop words e Pontuação**

```python
stops = stopwords.words("portuguese")
print(len(stops))
print(stops)
```

```python
palavras_sem_stopwords = [p for p in tokens if p not in stops]
print(texto)
print(palavras_sem_stopwords)
print("quantidade de palavras sem stopwords:",len(palavras_sem_stopwords))
print("quantidade de palavras com stopwords:",len(tokens))
```

muito útil para tratar textos em que a frequ^encia das palavras intefere nas técnicas

```python
##pontuação
print(string.punctuation)
```

```python
palavras_sem_pontuacao = [p for p in palavras_sem_stopwords if p not in string.punctuation]
print(texto)
print(palavras_sem_pontuacao)
print(len(palavras_sem_pontuacao))
```

**Métricas**


mostra uma fisribuição de frequência entre os tokens. Mostra as principais de forma ordenada.

```python
freq = nltk.FreqDist(palavras_sem_pontuacao)
freq
```

```python
mais_comuns = freq.most_common(5)
mais_comuns
```

**Stemming**

```python
#Porter Stemming
stemmer = PorterStemmer()
stem1 = [stemmer.stem(word) for word in palavras_sem_pontuacao]
print(palavras_sem_pontuacao)
print(stem1)
```

```python
stemmer2 = SnowballStemmer('portuguese')
stem2 = [stemmer2.stem(word) for word in palavras_sem_pontuacao]
print(palavras_sem_pontuacao)
print(stem2)
```

```python
stemmer3 = LancasterStemmer()
stem3 = [stemmer3.stem(word) for word in palavras_sem_pontuacao]
print(palavras_sem_pontuacao)
print(stem3)
```

o pos tag de nltk não suporta a lingua portuguesa..

```python
token2 = sent_tokenize(texto)

ntokens = []
for tokensentenca in token2:
  ntokens.append(word_tokenize(tokensentenca))
print(ntokens)

#pos tag feito a nivel de sentença
possentenca = pos_tag_sents(ntokens)
print(possentenca)
```

**Lemmatization**

```python
lemmatizer = WordNetLemmatizer()
resultado = [lemmatizer.lemmatize(palavra) for palavra in palavras_sem_pontuacao]
print(palavras_sem_pontuacao)
print(resultado)
```
